# Assignment 1
Stock value calculator:  100 shares of STOR capital stock was bought for $28.00 a piece and grows at 7% an year (average). It has a dividend yield of 3.4% and keeps increasing this yield by 50% each year. Write a python program which calculates (a) net worth of the stock after any given number of years (say, values like 3 years, 7 years, etcâ€¦) without dividend reinvestment, and (b) with dividend reinvestment. For explanations of dividend yield, go here : https://www.investopedia.com/terms/d/dividendyield.asp

### Function
This function takes in the inputed values and prints out the net worth after 'n' years depending on if the dividends were reinvested or not. It also prints out the fiscal gains, the increase in percentage, and the total dividends earned regardless of whether you reinvested them or not. 

```python
# Net worth function
def stock_calc(initial_worth, years, div):
    
    # Heading pt. 1
    heading = "Initial Investment: " + "${:,.2f}".format(initial_worth)
    
    # Heading pt. 2
    if div == True:
        print(heading, "with dividend reinvestment")
    else:
        print(heading, "without dividend reinvestment")
    
    # Set up variables
    total_div = 0
    net_worth = initial_worth
    
    # Iterate through the number of years, add 1 to the range to make it inclusive
    for i in range(1, years + 1):  
        
        # Get dividend yield factor
        div_rate = 0.034 * (1.5**(i-1))
        
        # Get dividends and new total
        div_cycle = div_rate * net_worth
        total_div += div_cycle
        
        if div == True:
            net_worth = net_worth * (1.07 + div_rate)
        else:
            net_worth = net_worth * 1.07
                
    # Final net worth
    print("Net Worth after", years, "years:", "${:,.2f}".format(net_worth))      
    
    # Profit
    profit = net_worth - initial_worth
    print("Your investment grew by", "${:,.2f}".format(profit))
    
    # Percent Change
    pct_change = "{:,.2f}%".format(profit / initial_worth * 100)
    print("Your investment increased by", pct_change)
    
    # Dividend reinvestment
    total_div = "${:,.2f}".format(total_div)
    
    # Print dividend amount
    if div == True:
        print("You reinvested", total_div, "in dividends")
    else:
        print("You earned", total_div, "in dividends")
```

### Set Up
This section of code gets the input from the user. It gets the number of years, and your decision whether to reinvest the dividends or not. Then it calls the stock_calc() function above so it can generate the information you need.

```python
# Get number of years and convert answer to integer type
y = input("Number of years you want to hold 100 shares of STOR capital stock purchased at $28.00 a piece: ")
y = int(y)

# Ask if user wants to consider dividend reinvestment
dividend = input("Would you like to consider dividend reinvestment? (Y/N): ")

# Convert yes/no to True/False
if dividend[0].lower() == 'y':
    dividend = True
elif dividend[0].lower() == 'n':
    dividend = False

# Initial Investment
shares = 100
price = 28
invest = shares * price

# Call function
stock_calc(invest, y, dividend)
```
