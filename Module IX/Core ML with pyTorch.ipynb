# It all starts with imports, as usual !

```python
'''Import Pytorch and if you have CUDA, take advantage of it.pssst: I dont!'''
import torch
import torch.optim as optim
import torch.nn as nn
import numpy as np

'''Do you have CUDA, Hmm what even is CUDA?'''
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

# One Tensor (array) to numpy them with

```python
'''Have you heard data is king?'''
np.random.seed(42)
x = np.random.rand(1000, 1)
y = 1 + 2 * x + .1 * np.random.randn(1000, 1)

'''No index will be left unrandomized oln my watch'''

idx = np.arange(1000)
np.random.shuffle(idx)

'''First 700 random indices for train'''

train_idx = idx[:700]

'''And the remaining indices for validation'''
val_idx = idx[700:]

'''Generates train and validation sets'''

x_train, y_train = x[train_idx], y[train_idx]
x_val, y_val = x[val_idx], y[val_idx]
```

# One tensor to pyTorch them with

```python

'''Our data was in Numpy arrays,  transform them into PyTorch's Tensors
and then we send them to the chosen device (As if I had a choice)'''

x_train_tensor = torch.from_numpy(x_train).float().to(device)
y_train_tensor = torch.from_numpy(y_train).float().to(device)

''' .type() tells us WHERE the tensor lives'''

print(type(x_train), type(x_train_tensor), x_train_tensor.type())
```

# And with one tensor train them !

```python
a = torch.randn(1, dtype=torch.float).to(device)
b = torch.randn(1, dtype=torch.float).to(device)

'''Require gradients, else you might as well use numpy'''

a.requires_grad_()
b.requires_grad_()
print(a, b)

'''When you add _, you will lose change them in place...'''
```

```python
'''Old friend learning rate is back !Epochs too...'''
lr = 1e-1
n_epochs = 1000

torch.manual_seed(42)
a = torch.randn(1, requires_grad=True, dtype=torch.float, device=device)
b = torch.randn(1, requires_grad=True, dtype=torch.float, device=device)
```

```python
'''And the loopy loop for updating the parameters...'''

for epoch in range(n_epochs):
    yhat = a + b * x_train_tensor
    error = y_train_tensor - yhat
    loss = (error ** 2).mean()
    
    '''Backwards from the loss'''
    loss.backward()
    
    '''Don't trust me, print the computed gradients...'''
    print(a.grad)
    print(b.grad)
    
    '''Parameters...update them...see below'''
        
    
    '''NO_GRAD keeps the update out of the gradient computation
    Why? Well, it's because of the dynamic graph from PyTorch'''
    
    with torch.no_grad():
        a -= lr * a.grad
        b -= lr * b.grad
    
    '''Let it go, let it gooo PyTorch, dont catch the computed gradients'''
    a.grad.zero_()
    b.grad.zero_()
    
print(a, b)
```
