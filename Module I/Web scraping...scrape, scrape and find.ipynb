# Scraping webpages for email addresses...psst...with more practice, we can turn hackers...

Import necessary libraries...aka...why re-invent the wheel when you can steal, er, share one.

```python
import urllib.request
import re
```

```python
# http://www.microstrategy.com/us/company/contact-us, let's form a "tunnel" to it.

my_page = urllib.request.urlopen("http://www.microstrategy.com/us/company/contact-us")
# What does it look like ? Non-html folks, please do not faint !

my_page_text = str(my_page.read())
```

```python
# Finally, time to find the needles in a haystack, I mean email addresses in a sea of text.
#For longer and more complex regular expression, you will need many more cups of coffee. Usually, in-person classes,
#coffee on me...

email_list = re.findall(">(\w*\S*\w*@\w*\.\w+)", my_page_text)
```

```python
print(email_list)
```
