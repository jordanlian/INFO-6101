# Them distances

```python
'''Very useful to data science chasers...let's start with the simplest, Euclidean distance'''
from scipy.spatial.distance import euclidean

p1 = (10,2) #point 1 coordinates
p2 = (1,0)  #point 2 coordinates

res = euclidean(p1, p2)
print(res)
```

```python
'''Manhattan distance...city block distance...useful in graph theory too'''

from scipy.spatial.distance import cityblock

p1 = (1, 0)
p2 = (10, 2)

res = cityblock(p1, p2)

print(res)
```

```python
'''Hamming distance...well talk binary'''

from scipy.spatial.distance import hamming

p1 = (True, False, True)
p2 = (False, True, True)

res = hamming(p1, p2)

print(res)
```
