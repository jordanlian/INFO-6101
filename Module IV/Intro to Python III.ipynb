# Python functions: what, why and how ?

Why functions ?

Modularize code

Reduce redundancy

What are functions ?

They are chunks of code that do something on their own !

How to write fucntions in python ?

```python
# Our first function to add two numbers

def add_nums(num1,num2):
    '''This function adds anything'''
    num3 = num1 + num2
    return num3
```

```python
add_nums(10,20)
```

```python
add_nums("Fred","Hutch")
```

```python
add_nums([1,2,3],[3,4,5])
```

```python
add_nums((1,2,3),(4,5))
```
